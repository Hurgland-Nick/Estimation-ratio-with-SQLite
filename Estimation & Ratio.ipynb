# Objective: Calculate an estimate of the population in 2050

### Import librairies

```python
# Import pandas library
import pandas as pd

# Import sqlite3 library
import sqlite3

# Import math library
import math

#from IPython.core.display import display
#pd.options.display.width = 0
```

### Database

```python
# Connection to the database
Connection = sqlite3.connect('factbook.db')
```

```python
# Connection to the database and execute query with pandas
a = pd.read_sql_query('select * from facts limit 10;', con = Connection)
display(a)
```

### Fomula to calculate an estimate of the population

```python
#Function
def pop_growth(x):
	return x['population'] * math.e **((x['population_growth']/100)*35)
  
# Apply calcul to each row by creating a new column
a['pop_2050'] = a.apply(lambda row: pop_growth(row), axis = 1)
display(a)
```

```python
# Delete NAN
a = a.dropna(axis = 0)
a = a[(a['area_land'] > 0) & (a['population']> 0)]
display(a)
```

```python
# Sort The result (ascending)
b = a.sort_values(by = ['pop_2050'], ascending = False)
display(a)
```

```python
# Display names
b = b['name'].iloc[0:9]
display(b)
```

# Purpose: to calculate the ratio of the total area of land to the total area of the ocean belonging to the countries of the table



```python
areaLand_areaWater = pd.read_sql_query('select SUM(area_land),SUM(area_water) from facts where area_land is not NULL;', Connection)
display(areaLand_areaWater)
```

```python
# Display ratio methode 1
ratio = pd.read_sql_query('select round(SUM(area_land)/SUM(area_water),1) from facts where area_land is not NULL;', Connection)
display(ratio)
```
